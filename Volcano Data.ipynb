# **Import Dataset**

```python
import pandas as pd
import numpy as np
import json
from datetime import datetime
```

# **Dataset**

```python
df = pd.read_csv("PetaBencana_Dataset.csv")
df
```

```python
df["tahun"] = df["date"].map(lambda x: datetime.strptime(x,'%d/%m/%Y').strftime("%Y"))

flood_depth = [json.loads(x) if str(x) != "nan" else str(0) for x in df["report_dat"].tolist()]
flood_depth_2 = [x["flood_depth"] if "flood_depth" in x else 0 for x in flood_depth ]
df["flood_depth"] = flood_depth_2

df_coba = df.query("Provinsi == 'Daerah Khusus Ibukota Jakarta' and disaster_type == 'flood'")
df_coba.head()

df_coba.groupby(["Kecamatan","tahun"]).agg({"flood_depth":[np.mean,np.median]}).to_excel("wkwkw.xlsx")
```
