```python
link = 'https://drive.google.com/open?id=1lDsrqy8xhNqQp2DiJ8bXJJBFdk1kocCw'
fluff, id = link.split('=')
print(id)
downloaded = drive.CreateFile({'id':id}) 
downloaded.GetContentFile("PetaBencana_Dataset.csv")  
df = pd.read_csv("PetaBencana_Dataset.csv")
df
```

```python
df1 = df.drop(columns=['pkey', 'source','status','url','image_url','title','tags','ID_Desa','ID_Kec','ID_Kab','ID_Prov']) #Drop Columns
df2 = df1[(df1['disaster_type']=="flood")]
df2.dropna()
df2.reset_index()
down = df2['long'].quantile(0.05)
up = df2['long'].quantile(0.95)
IQR = up - down    #IQR is interquartile range

filter = (df2['long'] >= down - 2*IQR ) & (df2['long'] <= up + 0.5 *IQR)
df3 = df2.loc[filter]  
df3.reset_index()
```

```python
down_ = df2['lat'].quantile(0.05)
up_ = df2['lat'].quantile(0.95)
IQR = up_ - down_    #IQR is interquartile range

filter = (df2['lat'] >= down_ - 1.5*IQR ) & (df2['lat'] <= up_ + 1.5 *IQR)
df3_ = df2.loc[filter]  
df3_.reset_index()
```

```python
from shapely.geometry import Point
import geopandas as gpd
from geopandas import GeoDataFrame
import matplotlib.pyplot as plt

geometry = [Point(xy) for xy in zip(df3_['long'],df3_['lat'])]
gdf = GeoDataFrame(df3_, geometry = geometry)

file_geo = '/content/gadm41_IDN_4.json'
df_geo = gpd.read_file(file_geo)

gdf.plot(ax=df_geo.plot(figsize=(15,15)), marker="o", color="red", markersize=6);
```
